##### Data Wrangling w/ Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
import seaborn as sns
```

```python
# setting up graph size
plt.rcParams['figure.figsize']=[15,5]
```

```python
# Getting Daily S&P 500 price
# will ideally edit wtih yfinance or alpaca or IEX when implementing to 
#Simple_stock_app.py

url = "https://raw.githubusercontent.com/Patrick-David/AMVR/master/spx.csv"
df = pd.read_csv(url,index_col='date', parse_dates=True)

#To model returns lets use pct_change
daily_ret = df['close'].pct_change()

#cleanning NA values

daily_ret.dropna(inplace=True)

daily_ret.head()
```

```python
# Resampel returns / month and take standard dev as measure of Vol
# Then annualize by nultiplying sqrt of number of periods (12)

monthly_annual = daily_ret.resample('M').std()*np.sqrt(12)

print(monthly_annual.head())

# Let's see major market event show up in the volatility
plt.plot(monthly_annual)
plt.axvspan('1987','1989',color='r',alpha=.5)
plt.axvspan('2008','2010',color='r',alpha=0.5)
plt.title('Monthly annualized vol - Black Monday and 2008 Financial Crysis highlighted')
labs= mpatches.Patch(color = 'red',alpha = .5, label = 'Black Monday & 08 Crash')
plt.legend(handles=[labs])
```

```python
# For ea year, rank each month based on volatility between 1-12
ranked = monthly_annual.groupby(monthly_annual.index.year).rank()
final = ranked.groupby(ranked.index.month).mean()

final.describe()
```

```python
#the final average result over 32 years
final
```

```python
#plot results for S$P AMVR. clearly october has the hgihest ave vol
# rank and december has the lowest. Mean of 6.45 is plotted

b_plot = plt.bar(x=final.index, height = final)
b_plot[9].set_color('g')
b_plot[11].set_color('r')
for i,v in enumerate(round(final,2)):
    plt.text(i+.8,1,str(v),color = 'black',fontweight='bold')
plt.axhline(final.mean(),ls='-',color ='k',label=round(final.mean(),2))
plt.title('Average Monthly Volatility Rank S&P 500 since 1986')
plt.legend()
plt.show()
```

```python
# Take absolute AMVR from the mean, we see Dec and Oct are the biggest
# absolute moves (Oct - Upside, Dec - Downside), w/ Dec being the greatest move

fin = abs(final-final.mean())
print(fin.sort_values())
Oct_value = fin[10]
Dec_value = fin[12]
print('Extreme Dec Value:', Dec_value)
print('Extrene Oct Value:', Oct_value)
```

```python
# As our null is that no seasonality exists or alternatively that the 
# month / day does not mater in terms of AMVR, we can shuffle 'date'
# labels.
# for simplicity, we will shuffle the 'daily' returns data, which has
# the same effect as shuffling 'month' labels

# generate null data

new_df_sim= pd.DataFrame()
highest_only = []

count = 0
n = 1000
for i in range(n):
    # sample sample size as dataset, drop timestap
    daily_ret_shuffle = daily_ret.sample(8191).reset_index(drop=True)
    daily_ret_shuffle.index = (pd.bdate_range(start = '1986-1-3',periods = 8191))
    # Double check wehre that 8191 value gets calculated
    monthly_annual = daily_ret_shuffle.resample('M').std()*np.sqrt(12)
    ranked = monthly_annual.groupby(monthly_annual.index.year).rank()
    sim_final = ranked.groupby(ranked.index.month).mean()
    new_df_sim = pd.concat([new_df_sim,sim_final],axis =1)
    # Recording highest AMVR for ea yaer for p-hacking
    maximum_month = max(sim_final)
    highest_only.append(maximum_month)
    
#Calculate absolute deviation in AMVR from the mean
all_months = new_df_sim.values.flatten()
mu_all_months = all_months.mean()

abs_highest = [abs(x-mu_all_months) for x in highest_only]
```

###### P - hacking

```python
abs_all_months = abs(all_months-mu_all_months)
```

```python
count = 0
for i in abs_all_months:
    if i > Dec_value: 
        count +=1
ans = count/len(abs_all_months)
print('p-value', ans)
```

```python
# count number of months in sim data where abs AMVR is > Dec
# Here we are comparing against all months
count = 0
for i in abs_all_months:
    if i> Dec_value:
        count +=1
ans = count/len(abs_all_months)
print('p_value:',ans)
```

```python
# Same but just considering highest abs AMVR for each of 1000 trials
count = 0
for i in abs_highest:
    if i>Dec_value:
        count+=1
ans=count/len(abs_highest)
print('p_value:',ans)
```
