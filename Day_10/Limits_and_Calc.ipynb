#### Understanding limits
Let's review the idea of limits: Since they are basically a function's value as it approaches a certain number meaning the limit

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')

def func1(x:float)->float:
    return x/x**2 #x/(x^2) = 1/x
X_low = np.linspace(-.1,-1e-3,100)
X_high = np.linspace(1e-3,.1,100)
fig,ax = plt.subplots()
ax.plot(X_low,[func1(x) for x in X_low],color = 'r')
_ = ax.plot(X_high,[func1(x) for x in X_high],color = 'g')
```

```python
def func2(x: float)-> float:
    return(x**2-1)/(x-1)

X_low=np.linspace(0,.9999,100)
X_high= np.linspace(1.0001,2,100)
fig,ax = plt.subplots()
ax.plot(X_low,[func2(x) for x in X_low],color = 'r')
_=ax.plot(X_high,[func2(x) for x in X_high], color = 'g')
```

```python
#Let's make a sin(x)/x function
import math as m

def func3(x:float)->float:
    return m.sin(x)/x
X_low=np.linspace(0,m.pi**2,100)
X_high= np.linspace(0,m.pi/2,100)
fig,ax = plt.subplots()
ax.plot(X_low,[func3(x) for x in X_low],color = 'r')
_=ax.plot(X_high,[func3(x) for x in X_high], color = 'g')
```
