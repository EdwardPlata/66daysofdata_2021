```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn
%matplotlib inline
import numpy as np
from scipy import stats

plt.rcParams.update({
    'font.size'           : 24.0,
    'axes.titlesize'      : 'large',
    'axes.labelsize'      : 'medium',
    'xtick.labelsize'     : 'medium',
    'ytick.labelsize'     : 'medium',
    'legend.fontsize'     : 'large',
})
```

```python
# Let's pretend our website logs show the follow number of visits for the past 7 days
visit_data = np.array([6,4,7,4,9,3,5])
```

```python
xbar = np.mean(visit_data)
s_var = np.var(visit_data)
print("Sample mean = {} \nSample variance= {}".format(xbar,s_var))
```

```python
#Let's see hits as a poisson dist
dist = stats.poisson(xbar)
print("Poisson(mu): p(X=0)={}".format(dist.pmf(0)))
```

```python
x = np.arange(dist.ppf(0.0001), dist.ppf(0.99))
fig,ax = plt.subplots(1,1,figsize=(12,5))
ax.plot(x,dist.pmf(x),'bo',ms=8, label = 'poisson pmf')
ax.vlines(x,0,dist.pmf(x),colors='b', lw=  5, alpha = 0.5)
ax.set_title("Poisson, $\lambda = {}".format(xbar));
ax.hist(visit_data, alpha=0.3, label='data histogram')
ax.legend();
```

```python
# flip a coin 100 times. Lets say we get 52 heads. What is the Method of Moments esto,ate  
```

```python
n = 100
p=52/n # we get 52 heads
dist = stats.binom(n,p)
print(dist.cdf(45))
```

```python
x = np.arange(dist.ppf(0.0001),
             dist.ppf(0.9999))
fig,ax= plt.subplots(1,1,figsize=(8,6))
ax.plot(x,dist.pmf(x),'bo',ms=8,label ='binom pmf')
ax.vlines(x,0,dist.pmf(x),colors='b',lw=5,alpha=0.5)
ax.set_title("Binomial PMF, n= 100, p = 0.52")
```

```python
uniform_dist = stats.uniform(0,50)
data = uniform_dist.rvs(10,random_state = 10)
print(data)
```

```python
xbar = np.mean(data)
2*xbar
```

```python
unif = stats.uniform(0,2*xbar)
plt.hist(data,alpha = 0.3)
x = np.linspace(-1,60,1000)
plt.plot(x, unif.pdf(x),'b',linewidth = 3);
```

#### What if we assume a normal distribution instead?

```python
normal = stats.norm(np.mean(data),np.std(data))
plt.hist(data,alpha=0.3)
x =np.linspace(-1,60,1000)
plt.plot(x,normal.pdf(x),'b',linewidth = 3);
```
