```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optim
import scipy.stats as scs
import pandas as pd


from itertools import product 

%matplotlib inline

plt.style.use('ggplot')
```

```python
normal = scs.norm(0,1)
fig,axs=plt.subplots(3,2,figsize=(16,6), sharex = True)

for ax in axs.flatten():
    samp = normal.rvs(20)
    ax.scatter(samp,np.repeat(0,20))
    mean = np.mean(samp)
    ax.scatter([mean],0,c='black',s=80)
```

The sample maximum is defined by m ax (X1,X2,...Xn)

```python
normal = scs.norm(0,1)
fig,axs=plt.subplots(3,2,figsize=(16,6),sharex = True)

for ax in axs.flatten():
    samp = normal.rvs(20)
    ax.scatter(samp,np.repeat(0,20))
    M=np.max(samp)
    ax.scatter([M],0,c='black',s=80)
```

```python
normal = scs.norm(0,1)
fig,axs=plt.subplots(3,2,figsize=(16,6),sharex = True)
means=list()
for ax in axs.flatten():
    samp = normal.rvs(20)
    ax.scatter(samp,np.repeat(0,20))
    mean = np.mean(samp)
    means.append(mean)
    ax.scatter([M],0,c='black',s=80)
```

Distribution of the black points in the sampleing distribution of the mean

```python
fig,ax = plt.subplots(1,1,figsize=(16,2))
ax.scatter(means,np.repeat(0,6),c='black',s=50);
```

Ifwe superimpose a sample from the original distrbution, it's cery clear that the stpread of the sampeling distribution is much smaller than that of the original.

```python
fig, ax=  plt.subplots(1,1,figsize=(16,2))
ax.scatter(normal.rvs(20),np.repeat(0,20))
ax.scatter(means,np.repeat(0,6),c='black',s=50);
```
